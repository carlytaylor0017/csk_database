```python
df = pd.read_csv('version.smi.gz', compression='gzip', header=0, sep=' ', quotechar='"') 
```

```python
df.index = df['isosmiles'].str.len()
df = df.sort_index(ascending=True).reset_index(drop=True)
df.head()
```

```python
df.tail()
```

```python
df.count()
```

```python
counted = Counter(df['isosmiles'].str.cat())
element_df = pd.DataFrame([counted])
element_df = element_df.T
element_df.head()
```

```python
element_df.columns = ['freq']
element_df.head()
```

```python
element_df['symbol'] = element_df.index
element_df.reset_index(level=0, inplace=True)
element_df.head()
```

```python
del element_df['index']
element_df.head()
```

```python
element_df = element_df.sort_values('freq', ascending = True)
element_df.reset_index(level=0, inplace=True)
element_df.head()
```

```python
del element_df['index']
element_df
```

```python
element_df_top = element_df.iloc[0:32]
ax = element_df_top.plot.bar(x='symbol', y = 'freq', color="#dd55ff", position=0, width=0.5, rot=0)
```

```python
element_df_bottom = element_df.iloc[-36:]
ax = element_df_bottom.plot.bar(x='symbol', y = 'freq', color="#dd55ff", position=0, width=0.5, rot=0)
```

```python
element_list = list(element_df['symbol'])
element_list 
```

```python
element_list = [''.join(c for c in s if c not in string.punctuation) for s in element_list]
element_list = [''.join(c for c in s if c not in '1,2,3,4,5,6,7,8,9,0') for s in element_list]
element_list = list(filter(None, element_list))
element_list
```

```python
hc_df = df[(df['isosmiles'].str.contains('H')) 
           & (df['isosmiles'].str.contains('C')) 
           | (df['isosmiles'].str.contains('c'))]
hc_df.head()
```

```python
hc_df.tail()
```

```python
hc_df.count()
```

```python
hc_df = hc_df[(~hc_df['isosmiles'].str.contains('O')) 
              & (~hc_df['isosmiles'].str.contains('N')) 
              & (~hc_df['isosmiles'].str.contains('B')) 
              & (~hc_df['isosmiles'].str.contains('Br')) 
              & (~hc_df['isosmiles'].str.contains('F')) 
              & (~hc_df['isosmiles'].str.contains('P')) 
              & (~hc_df['isosmiles'].str.contains('Cl')) 
              & (~hc_df['isosmiles'].str.contains('Ca')) 
              & (~hc_df['isosmiles'].str.contains('S')) 
              & (~hc_df['isosmiles'].str.contains('N')) 
              & (~hc_df['isosmiles'].str.contains('Cs')) 
              & (~hc_df['isosmiles'].str.contains('Hf'))
              & (~hc_df['isosmiles'].str.contains('n'))
              & (~hc_df['isosmiles'].str.contains('f'))
              & (~hc_df['isosmiles'].str.contains('I'))
              & (~hc_df['isosmiles'].str.contains('s'))
              & (~hc_df['isosmiles'].str.contains('o')) 
              & (~hc_df['isosmiles'].str.contains('d')) 
              & (~hc_df['isosmiles'].str.contains('t')) 
              & (~hc_df['isosmiles'].str.contains('e')) 
              & (~hc_df['isosmiles'].str.contains('l')) 
              & (~hc_df['isosmiles'].str.contains('i')) 
              & (~hc_df['isosmiles'].str.contains('u')) 
              & (~hc_df['isosmiles'].str.contains('a')) 
              & (~hc_df['isosmiles'].str.contains('p')) 
              & (~hc_df['isosmiles'].str.contains('m')) 
              & (~hc_df['isosmiles'].str.contains('g')) 
              & (~hc_df['isosmiles'].str.contains('z'))
              & (~hc_df['isosmiles'].str.contains('r'))
              & (~hc_df['isosmiles'].str.contains('b'))
              & (~hc_df['isosmiles'].str.contains('y'))]
hc_df.head()
```

```python
hc_df.isosmiles.unique()
```

```python
hc_unique = hc_df['isosmiles'].str.replace('[{}]'.format(string.punctuation), '').str.replace(r'\[.*?\]|\(.*?\)|\W', '').str.replace('\d+', '')
hc_unique_2 = hc_unique.apply(lambda x: x.lower())
hc_unique_2.unique()
```

```python
unique_str = []
for word in hc_unique_2:
    for i in word:
        if i not in unique_str:
            unique_str.append(i)
```
